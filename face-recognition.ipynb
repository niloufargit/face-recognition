```python
import cv2
```

```python
#step1 : read image
woman1_path = 'train-model/woman-1.jpg'

img = cv2.imread(woman1_path)
```

```python
img.shape
```

```python
#step2 :  image to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gray_image.shape
```

```python
# object detection via classifier 
# haarcascade_frontalface_default.xml : a pre-trained classifier that is used to detect frontal faces in an image or video stream
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

```python
# face detection 
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
)
```

- gray_image: This is the grayscale version of the image in which we want to detect faces.
 
- scaleFactor: This parameter compensates for the fact that faces in an image may appear at different scales. A value of 1.1 means that the algorithm will look for faces that are 10% smaller than the previous scale.
 
- minNeighbors: This parameter specifies how many neighbors each candidate rectangle should have to retain it. A higher value results in fewer detections but with higher quality.
 
- minSize: This parameter specifies the minimum size of the detected face. Any face smaller than this size will be ignored.
